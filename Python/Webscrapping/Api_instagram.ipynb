```python
from instagrapi import Client
import pandas as pd
```

### Autenticaci贸n para entrar a Instagram

```python
#M茅todo de requisici贸n de informaci贸n informando que entraremos como cliente
cl = Client()
#M茅todo de autenticaci贸n con usuario y contrase帽a
cl.login('', '')
```

### Solicitud de informaci贸n a la API

```python
# Se ingresa la informaci贸n que queremos obtener con el nombre especifico del #hashtag.
hasgtag = 'messi'

#Ingresa el requerimiento a la API.
#la variable amount indica la cantidad de publicaciones que el requerimiento tendr谩.
medias = cl.hashtag_medias_top(hasgtag, amount=2)

#El output obtenido lo convertimos a un dictionario y para fines practicos, solamente usaremos el primer argumento
dict_htag = medias[0].dict()
dict_htag
```

### Limpieza de datos

Viendo el dictionario de una celda anterior, podemos ver que:
- 'users' y 'resources' tienen como valor otro dictionario.
- El ultimo valor del dictionario (clips_metadata) no da informaci贸n relevante.
- Dentro de una misma publicaci贸n se puede agregar multiple fotos y/o videos, por lo que 'resources' nos brinda la informaci贸n individualmente de cada una de ellas (Es la informaci贸n de aquel contenido que no sea la primera foto y/o video en la publicaci贸n).

```python
if dict_htag['resources']==[]:
    dict_htag['resources']
else:
    # Obtenemos el largo del valor 'resources' que contiene el contenido de la publicaci贸n
    len(dict_htag['resources'])

    # Imprimimos para tener una comparativa del filtrado de informaci贸n antes y despues
    print(dict_htag['resources'][1])

    # Se quita la informaci贸n de los valores 'pk', 'video_url' y 'media_type' para solamente quedarnos con
    # el valo de 'thumbnail_url' que es la liga url del contenido de publicaci贸n ya sea foto y/o video
    for x in range(len(dict_htag['resources'])):
        dict_htag['resources'][x] = {'thumbnail_url':dict_htag['resources'][x]['thumbnail_url']}

    # Imprimimos nuevamente para poder ver si el valor del dict 'resources' solamente muestra la info de 'thumbnail_url'
    print(dict_htag['resources'][1])
```

```python
### Optamos por quitar el ultimo valor del diccionario

# Identificamos los valores
print(dict_htag.keys())
print(' ')

# Se vacian los nombres del diccionario en una lista, omitiendo los ultimos dos valores
key_list = list(dict_htag.keys())[:-1]

# Se filtra el dictionario con los valores del listado key_list y se hace una comparativa para confirmar con el print
dict_htag = dict((k, dict_htag[k]) for k in key_list)
print(dict_htag.keys())
```

```python
# Segun la documentaci贸n de la funci贸n pandas el json_normalize, nos ayuda a que un dato semi estructado de JSON
#se convierta en dataframe
df_htag = pd.json_normalize(dict_htag, sep='_')
```

```python
# Imprimimos la columna para poder identificar el nombre de las columnas resultantes
df_htag.columns
```

Notamos que:
- Las columnas que tienen 'user_' al principio, son los valores del dictionario dentro de la llave 'users'. La funci贸n pd_json_normalize nos ayudo relacionar el nombre principal y el nombre secundario para que el valor de dicha relaci贸n se pueda  asignar a una celda del dataframe.
<br />
- El valor 'resources' al estar relacionado a ls publicaciones de contenido, puede tener una o m谩s fotos y/o videos, por lo que el valor del dictionario son varios dictionarios y el pd.json_normalize en lugar de actuar como con la llave 'users' se guarda todo el contenido como lista en el dataframe.

```python
# La columna resources la pasamos a otro dataframe y cada valor del listado se almacena en una columna
# Esto nos da como resultado que cada valor de la celda del nuevo dataframe sea un dictionario
# con el valor de 'thumbnail_url' por cada publicaci贸n

#Nota, en caso de que salga un dataframe nulo, es que la publicaci贸n solamente tiene una imagen/video y esta
#ya se encuentra guardada. Dado que estos pasos solamente son para cuando tiene m谩s de una foto y/o video la publicaci贸n
df_htag2 = pd.DataFrame(df_htag['resources'].tolist()) \
                                            .add_prefix('contenido_')

df_htag2
```

```python
# Mediante un for para la nueva base aplicamos un apply pd.Series para que el valor del dictionario se guarde
# en la celda correspondiente. Esto nos dar谩 que el valor de cada uno de la celda sea la liga del contenido de
# la publicaci贸n

for x in range(df_htag2.shape[1]):
    df_htag2[str('contenido_'+str(x))] = df_htag2[str('contenido_'+str(x))].apply(pd.Series)

df_htag2
```

```python
# Al ser una extracci贸n la base df_htag2 de la original tiene la misma dimensi贸n asi que un pd.concat
# se anexa en el mismo orden que el index, siempre y cuando no se hayan modificado las bases
df_htag_final = pd.concat([df_htag, df_htag2], axis=1)
```
