**Objetivo**:
Crear una función para ejecutar un comando de python en donde en caso de ejecutarse
de manera efectiva se verá el resultado y para el caso contrario hará lo siguiente:

    - Imprimira el error por el cual no se realizo de manera correcta.
    - Abrira una nueva pagina de google.
    - En el buscador colocara el error de la función.
    - Seleccionara el primer link de busqueda de la pagina de stack over flow.
    - Imprimira el comentario más votado.

# Script

```python
# Librerias necesarias

import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
from selenium.webdriver.support.ui import WebDriverWait
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
```

```python
#Función para solicitar ayuda de Stack Over Flow
def get_SOS_help(command):

    #Se intenta la función que funcione para que la realice correctamente
    try:
        command = 'print(' + command + ')'
        return exec(command)

    #En caso que encuentre error, imprime el error, lo busca en google e imprime el primer link con el apoyo más votado
    except Exception as error: #guarda el error obtenido en la variable 'error'
        #Imprime el error del comando
        print('The command cant be performe: '+ '['+str(error)+']')

        #Se conecta al driver de GoogleChrome
        driver = webdriver.Chrome(ChromeDriverManager().install())
        WAIT_TIME = 10
        driver.wait = WebDriverWait(driver, WAIT_TIME)

        #Le damos la url de goole para seleccionar que a esta página es a la que nos vamos a conectar.
        url = 'http://www.google.com'
        driver.get(url)

#       #Tiempo que le indicamos al driver para que espere
#         time = 5
        #Le damos click al boton de aceptar cookies de la pagina de google
        cookies_buttom = driver.find_element("css selector", 'button#L2AGLb.tHlp8d')
        cookies_buttom.click()

        #Insertamos el error en el buscado de google
        search_err = driver.find_element("css selector", 'input.gLFyf') #Encuentra la barra del buscador de google
        search_err.send_keys(str(error)) #Inserta el error en el buscador
        search_err.send_keys(Keys.RETURN) #Click para proceder a la navegación en el buscador

        #Imprimir primer link
        html_doc = driver.page_source #obtenemos el nuevo link procedente de la busqueda del error
        soup = BeautifulSoup(html_doc, 'html.parser') #convertimos la url en objeto soup
        results = soup.find("div", {"class": "yuRUbf"}) #metodo webscrapping para imprimir la primer liga
        results2 = results.find("a", href=True) #metodo webscrapping para imprimir la primer liga
        print('first link:')
        print (results2['href'])


        #Navegación en google con la URL nueva que es la busqueda del error
        first_link = driver.find_element("css selector", 'h3.LC20lb.MBeuO.DKV0Md') #Seleccionamos el primer link
        first_link.click() #Click acceder a la primera opción de página web

        #Le damos click al boton de aceptar cookies de la pagina stack over flow
        cookies_buttom_stack = driver.find_element("css selector", 'button.flex--item.s-btn.s-btn__primary.js-accept-cookies.js-consent-banner-hide')
        cookies_buttom_stack.click()

        print("")
        print("")
        print("")


        #Comentario más votado
        html_doc2 = driver.page_source #obtenemos el nuevo link de la navegación que es de StackOverFlow
        soup_file = BeautifulSoup(html_doc2, 'html.parser') #convertimos la url en objeto soup
        results3 = soup_file.find("div", {"class": "s-prose js-post-body"}) #método webscrapping para la primer liga
        results4 = results3.find("code") #método webscrapping para imprimir la primer liga
        print('Most voted help:')
        print('')
        print(results4.text)

#         print(stack_overflow_first_link)
#         print(stack_overflow_most_voted_help)
#         prompt_web(stack_overflow_first_link)
    return
```
